```python
import py_entitymatching as em
path_A = 'advisor_clean.csv'
path_B = 'yelp_clean.csv'
```

```python
A = em.read_csv_metadata(path_A, key='ID')
B = em.read_csv_metadata(path_B, key='ID')
```

```python
# A
```

```python
ab = em.AttrEquivalenceBlocker()
C0 = ab.block_tables(A, B, 'postal code', 'postal code',
                    l_output_attrs=['shop name', 'postal code', 'street address'], 
                    r_output_attrs=['shop name', 'postal code', 'street address'], allow_missing=True)
ob = em.OverlapBlocker()
C0 = ob.block_candset(C0, 'shop name', 'shop name', word_level=True, overlap_size=1)
C0 = ob.block_candset(C0, 'street address', 'street address', word_level=True, overlap_size=3)
C0
```

```python
ab = em.AttrEquivalenceBlocker()
C1 = ab.block_tables(A, B, 'postal code', 'postal code',
                    l_output_attrs=['shop name', 'postal code', 'street address'], 
                    r_output_attrs=['shop name', 'postal code', 'street address'], allow_missing=True)
ob = em.OverlapBlocker()
C1 = ob.block_candset(C1, 'shop name', 'shop name', word_level=True, overlap_size=2)
C1 = ob.block_candset(C1, 'street address', 'street address', word_level=True, overlap_size=2)
C1
```

```python
ab = em.AttrEquivalenceBlocker()
C2 = ab.block_tables(A, B, 'phone number', 'postal code',
                    l_output_attrs=['shop name', 'postal code', 'street address'], 
                    r_output_attrs=['shop name', 'postal code', 'street address'], allow_missing=True)
ob = em.OverlapBlocker()
C2 = ob.block_candset(C2, 'shop name', 'shop name', word_level=True, overlap_size=3)
C2 = ob.block_candset(C2, 'street address', 'street address', word_level=True, overlap_size=2, allow_missing=True)
C2
```

```python
ab = em.AttrEquivalenceBlocker()
C3 = ab.block_tables(A, B, 'phone number', 'phone number',
                    l_output_attrs=['shop name', 'phone number', 'street address'], 
                    r_output_attrs=['shop name', 'phone number', 'street address'], allow_missing=True)
ob = em.OverlapBlocker()
C3 = ob.block_candset(C3, 'shop name', 'shop name', word_level=True, overlap_size=3)
C3 = ob.block_candset(C3, 'street address', 'street address', word_level=True, overlap_size=2, allow_missing=True)
C3
```

```python
ob = em.OverlapBlocker()
C4 = ob.block_tables(A, B, 'shop name', 'shop name',
                    l_output_attrs=['shop name', 'phone number', 'street address'], 
                    r_output_attrs=['shop name', 'phone number', 'street address'], word_level=True, overlap_size=2)
C4 = ob.block_candset(C4, 'street address', 'street address', word_level=True, overlap_size=3)
C4
```

```python
ab = em.AttrEquivalenceBlocker()
C5 = ab.block_tables(A, B, 'street address', 'street address',
                    l_output_attrs=['shop name', 'phone number', 'street address'], 
                    r_output_attrs=['shop name', 'phone number', 'street address'])
C5 = ob.block_candset(C5, 'shop name', 'shop name', word_level=True, overlap_size=1)
C5
```

```python
C = em.combine_blocker_outputs_via_union([C0, C1, C2, C3, C4, C5])
C
```

```python
# now debug the blocker output
dbg = em.debug_blocker(C, A, B, output_size=100)
dbg
```

```python
em.to_csv_metadata(C, './candidate_set.csv')
```
