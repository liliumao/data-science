```python
import py_entitymatching as em
path_A = '/home/luobo/codeLibrary/data_science/advisor_clean.csv'
path_B = '/home/luobo/codeLibrary/data_science/yelp_clean.csv'
```

```python
A = em.read_csv_metadata(path_A, key='ID')
B = em.read_csv_metadata(path_B, key='ID')
```

```python
A
```

```python
ab = em.AttrEquivalenceBlocker()
C0 = ab.block_tables(A, B, 'phone number', 'phone number',
                    l_output_attrs=['shop name', 'phone number'], 
                    r_output_attrs=['shop name', 'phone number'], allow_missing=True)
```

```python
C0
```

```python
ob = em.OverlapBlocker()
C1 = ob.block_candset(C0, 'shop name', 'shop name', word_level=True, overlap_size=1)
C1
```

```python
# now debug the blocker output
dbg = em.debug_blocker(C1, A, B, output_size=100)
dbg
```

```python
em.to_csv_metadata(C1, './candidate_set.csv')
```
