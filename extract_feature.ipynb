```python
import py_stringmatching as sm
import csv
```

```python
with open("candidate_set.csv", 'rb') as csvfile:
    reader = csv.DictReader(csvfile)
    l_name = []
    r_name = []
    l_postalc = []
    r_postalc = []
    l_addr = []
    r_addr = []
    l_phone = []
    r_phone = []
    length = 0
    for row in reader:
        l_name.append(row['ltable_shop name'].split(" "))
        l_postalc.append(row['ltable_postal code'])
        l_addr.append(row['ltable_street address'].split(" "))
        l_phone.append(row['ltable_phone number'])
        r_name.append(row['rtable_shop name'].split(" "))
        r_postalc.append(row['rtable_postal code'])
        r_addr.append(row['rtable_street address'].split(" "))
        r_phone.append(row['rtable_phone number'])
        length += 1
    
    name_tfidf = sm.similarity_measure.tfidf.TfIdf(l_name + r_name)
    addr_tfidf = sm.similarity_measure.tfidf.TfIdf(l_addr + r_addr)
    ret = []
    
    for i in xrange(length):
        res = {}
        
        if not l_phone[i] or not r_phone[i]:
            res['phone'] = 0
        elif l_phone[i] == r_phone[i]:
            res['phone'] = 1
        else:
            res['phone'] = -1
        
        if not l_postalc[i] or not r_postalc[i]:
            res['post'] = 0
        elif l_postalc[i] == r_postalc[i]:
            res['post'] = 1
        else:
            res['post'] = -1
        
        res['name'] = name_tfidf.get_sim_score(l_name[i], r_name[i])
        res['addr'] = addr_tfidf.get_sim_score(l_addr[i], r_addr[i])
        
        ret.append(res)
```
