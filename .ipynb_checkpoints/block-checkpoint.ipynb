```python
import py_entitymatching as em
path_A = 'advisor_clean.csv'
path_B = 'yelp_clean.csv'
A = em.read_csv_metadata(path_A, key='ID')
B = em.read_csv_metadata(path_B, key='ID')
```

```python
ab = em.AttrEquivalenceBlocker()
C0 = ab.block_tables(A, B, 'postal code', 'postal code',
                    l_output_attrs=['shop name', 'postal code', 'street address'], 
                    r_output_attrs=['shop name', 'postal code', 'street address'], allow_missing=True)
```

```python
ob = em.OverlapBlocker()
C1 = ob.block_candset(C0, 'shop name', 'shop name', word_level=True, overlap_size=1)
C1
```

```python
ob = em.OverlapBlocker()
C2 = ob.block_candset(C1, 'street address', 'street address', word_level=True, overlap_size=2, allow_missing=True)
C2
```

```python
dbg = em.debug_blocker(C2, A, B, output_size=100)
dbg
```

```python
em.to_csv_metadata(C2, './candidate_set.csv')
```
